```
import pandas as pd
!pip install yfinance
import yfinance as yf
import datetime
from datetime import date, timedelta
today = date.today()
```

```
#In the above code, I have collected the latest data of Bitcoin prices
#for the past 730 days, d1 = today.strftime("%Y-%m-%d")
d1 = today.strftime("%Y-%m-%d")
end_date = d1
d2 = date.today() - timedelta(days=730)
d2 = d2.strftime("%Y-%m-%d")
start_date = d2
d1,d2
```

```
data = yf.download('BTC-USD', 
                      start=start_date, 
                      end=end_date, 
                      progress=False)
print(data)
```

```
data["Date"] = data.index
data = data[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
data.reset_index(drop=True, inplace=True)
print(data.head())
```

```
#change in bitcoin prices till today by using a candlestick chart:

!pip install plotly
import plotly.graph_objects as go
figure = go.Figure(data=[go.Candlestick(x=data["Date"],
                                        open=data["Open"], 
                                        high=data["High"],
                                        low=data["Low"], 
                                        close=data["Close"])])
figure.update_layout(title = "Bitcoin Price Analysis", 
                     xaxis_rangeslider_visible=False)
figure.show()
```

```
# Predicting the future closing prices of cryptocurrency is our target variable. 
# The AutoTS library in Python is one of the best libraries for time series analysis. 
# So here I will be using the AutoTS library to predict the bitcoin prices for the next 30 days:
!pip install autots
from autots import AutoTS
```

```
model = AutoTS(forecast_length=30, frequency='infer', ensemble='simple')
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```
